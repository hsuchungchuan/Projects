```python

# file name and path
file_path = r"Z:\AION\Data\2024\12\05\mot_ndscan\000017872-MeasureRedMOT.h5"
# Open the HDF5 file to extract the relevant datasets
with h5py.File(file_path, 'r') as h5file:
    # Extracting the axis and data for plotting
    x1 = h5file['datasets/ndscan.rid_17872.points.axis_0'][:]
    y1 = h5file['datasets/ndscan.rid_17872.points.channel_red_mot_normalised_bg_corrected_img_sum_horiz'][:]

x1 = np.delete(x1, 4)
y1 = np.delete(y1, 4)


# Plotting the data#
plt.figure(figsize=(12, 10))

plt.plot(x1/1e6, y1/10, 'o-', markersize = 20)  # Large dots and connected points



# # plt.title('nb red mot 0A normalized pixel sum vs dpaom frequency\n mirny 460MHz mod width 10MHz +2.0A for axial', fontsize=20, pad=20)
# # Increase tick label size
plt.tick_params(axis='both', which='major', labelsize=20)
plt.xlabel('Red DPAOM frequency (MHz)', fontsize=20)
plt.ylabel('Normalized horizontal pixel sum', fontsize=20)
plt.legend(fontsize=20)
ax2 = plt.twiny()
frequencyoffsetfromcavity = x1/1e6 *2 + 460 - 100
ax2.plot(frequencyoffsetfromcavity, y1/10, 'o-', alpha=0)  # Large dots and connected points
ax2.set_xlabel('Frequency offset from cavity (MHz)', fontsize=20, labelpad=12)
ax2.tick_params(axis='x', which='major', labelsize=20)

plt.title('Broadband red MOT spectroscopy', fontsize=25, pad=12)

# #plt.xlim(340, 360)
# # plt.ylim(-0.05,0.95)
plt.grid(True)
# # Adjust layout to prevent cut off
plt.tight_layout()
# # Save the plot with 300 DPI
# plt.savefig('241205 Broadband red MOT spectroscopy 1.pdf', dpi=1200)

plt.show()
```
