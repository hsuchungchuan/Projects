```python
file_1 =r'shelvingthirdapril\MirnyStart212MHz434829221THzEnd215MHzCSVs\MirnyStart212MHz434829221THzEnd215MHzCSVs_1.csv'
file_2 =r'shelvingthirdapril\MirnyStart212MHz434829221THzEnd215MHzCSVs\MirnyStart212MHz434829221THzEnd215MHzCSVs_2.csv'

shelving_5 = Shelving(file_1, file_2)
shelving_5.plot_shelving_graph(figure_size=(50, 5))

time_interval_5 = np.linspace(0, 33, 5)
frequency_lists_5 = range(221, 225)

average_amplitudes_5, standard_deviations_5 = \
    shelving_5.get_shelving_amplitude(time_interval_5, frequency_lists_5, 
                                      method = 'subtraction', peak_distance=700000, 
                                      upper_border=360, lower_border=360, 
                                      normalised=True)

```

```python
# Concatenating the data
total_average_amplitudes = np.concatenate((average_amplitudes_1, average_amplitudes_2, 
                                     average_amplitudes_3, average_amplitudes_4, 
                                     average_amplitudes_5))
total_standard_deviations = np.concatenate((standard_deviations_1, standard_deviations_2,
                                             standard_deviations_3, standard_deviations_4,
                                             standard_deviations_5))
total_frequency_lists = np.concatenate((frequency_lists_1, frequency_lists_2,
                                        frequency_lists_3, frequency_lists_4,
                                        frequency_lists_5))
print(total_frequency_lists)
print(total_average_amplitudes)
print(total_standard_deviations)

total_frequency_lists = np.delete(total_frequency_lists, [-1, -2, -3, 3, -4, -6, -5])
total_average_amplitudes = np.delete(total_average_amplitudes, [-1, -2, -3, 3, -4, -6, -5])
total_standard_deviations = np.delete(total_standard_deviations, [-1, -2, -3, 3, -4, -6, -5])
total_frequency_lists[0] = np.mean([total_frequency_lists[-4], total_frequency_lists[-6], 
                                    total_frequency_lists[0]])
plt.errorbar(total_frequency_lists, total_average_amplitudes, 
             yerr=total_standard_deviations, fmt='o')
for x, y in zip(total_frequency_lists, total_average_amplitudes):
    plt.text(x, y, str(x))
plt.xlabel('Frequency (MHz) + 434.829THz')
plt.ylabel('Normalised Amplitude')
plt.title('Shelving Amplitude against Frequency')
```
