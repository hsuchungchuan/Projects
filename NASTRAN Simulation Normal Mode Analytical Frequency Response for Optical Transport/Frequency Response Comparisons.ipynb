```python
df2 = pd.read_excel('Z:/Shared/AION/Data/Transport/Simulation (NASTRAN)/Frequency Response 210122.xlsx')
```

```python
onelens_freq = df2.iloc[:,0].values
onelens_disp = df2.iloc[:,1].values
onelens_thick_freq = df2.iloc[:,2].values
onelens_thick_disp = df2.iloc[:,3].values
threelens_freq = df2.iloc[:,4].values
threelens_disp = df2.iloc[:,5].values
threelens_thick_freq = df2.iloc[:,6].values
threelens_thick_disp = df2.iloc[:,7].values
```

```python
plt.figure(figsize=(12,8))
plt.plot(onelens_freq,onelens_disp,'x--',label='one lens normal rod')
plt.plot(onelens_thick_freq,onelens_thick_disp,'o--',label='one lens thick rod')
plt.plot(threelens_freq,threelens_disp,'.--',label='three lens normal rod')
plt.plot(threelens_thick_freq,threelens_thick_disp,'^--',label='three lens thick rod')
plt.title('Comparison between four different periscope systems (Nastran simulation modal frequency sweep)')
plt.xlabel('Frequency / Hz')
plt.ylabel('Maximum Displacement Magnitude / arb.')
plt.legend()
plt.show()
```

```python
# displacement of periscope due to a external force of 10N, simulated in Nastran (Linear Static Analysis)

delta_a = 0.1018 #one lens normal rod
delta_b = 0.0187 #one lens thick rod
delta_c = 0.0407 #three lens normal rod
delta_d = 0.0106 #three lens thick rod
```

```python
# see normal mode maximum maximum displacement magnitude (arb.) for each system

a = np.max(onelens_disp)
b = np.nanmax(onelens_thick_disp)
c = np.nanmax(threelens_disp)
d = np.nanmax(threelens_thick_disp)
```

```python
a,b,c,d
```

```python
# Comparison permutations:

# max normal mode displacements (arb)
r1 = a/b #one lens vs one lens thick
r2 = a/c #one lens vs three lens
r3 = b/d #one lens thick vs three lens thick
r4 = c/d #three lens vs three lens thick

# force displacements
d1 = delta_a/delta_b
d2 = delta_a/delta_c
d3 = delta_b/delta_d
d4 = delta_c/delta_d
```

```python
r1,r2,r3,r4,d1,d2,d3,d4
```

```python
d1/r1,d2/r2,d3/r3,d4/r4
```

```python
plt.plot([d1/r1,d2/r2,d3/r3,d4/r4],'x')
plt.show()
```

```python
deltas = [delta_a,delta_b,delta_c,delta_d]
displacements = [a,b,c,d]
```

```python
ratios = []
for i in range(4):
    for j in range(4):
        if i!=j:
            a = displacements[i]/displacements[j]
            b = deltas[i]/deltas[j]
            ratios.append(b/a)
```
