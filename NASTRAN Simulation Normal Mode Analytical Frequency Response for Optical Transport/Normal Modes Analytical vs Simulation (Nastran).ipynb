```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import time as time
plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 18
plt.rcParams['legend.fontsize'] = 18
plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['ytick.direction'] = 'out'
plt.rcParams['xtick.major.size'] = 5.0
plt.rcParams['xtick.minor.size'] = 3.0
plt.rcParams['ytick.major.size'] = 5.0
plt.rcParams['ytick.minor.size'] = 3.0
plt.rcParams['legend.handlelength'] = 1.0
```

```python
a = [10,15,12.5,17.5,7.5]
mode1_analytic = [np.sqrt((5-np.sqrt(19))*9.81*1000/i) for i in a]
mode1_simulation = [17,7.53,10.8,5.6,30.1]
mode2_analytic = [np.sqrt((5+np.sqrt(19))*9.81*1000/i) for i in a]
mode2_simulation = [231.7,102.7,148.3,76,408.1]
```

```python
plt.figure(figsize=(7.5,6))
plt.plot(a,mode1_analytic,'x',label='analytic solution')
plt.plot(a,mode1_simulation,'.',label='nastran simulation')
plt.legend()
plt.xlabel('Length of rod/string $a$ (mm)')
plt.ylabel('Normal mode frequency $\omega$ (Hz)')
plt.title('Normal Mode 1')
# plt.show()
plt.savefig('rodstring1'+'.pdf',bbox_inches='tight',dpi=300)
```

```python
plt.figure(figsize=(7.5,6))
plt.plot(a,mode2_analytic,'x',label='analytic solution')
plt.plot(a,mode2_simulation,'.',label='nastran simulation')
plt.legend()
plt.xlabel('Length of rod/string $a$ (mm)')
plt.ylabel('Normal mode frequency $\omega$ (Hz)')
plt.title('Normal Mode 2')
# plt.show()
plt.savefig('rodstring2'+'.pdf',bbox_inches='tight',dpi=300)
```

# Example: A uniform rod of length a hangs vertically on the end of a light inelastic string of length a, the string being attached to the upper end of the rod. What are the frequencies, and shapes, of the normal modes of small oscillations in a vertical plane?

Analytical solution: $\omega^2 = (5\pm\sqrt{19})\frac{g}{a}$, where $a$ is the length of the rod/string.

This implies that $\omega$ scales with $\frac{1}{\sqrt{a}}$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 15
plt.rcParams['legend.fontsize'] = 18
plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['ytick.direction'] = 'out'
plt.rcParams['xtick.major.size'] = 5.0
plt.rcParams['xtick.minor.size'] = 3.0
plt.rcParams['ytick.major.size'] = 5.0
plt.rcParams['ytick.minor.size'] = 3.0
plt.rcParams['legend.handlelength'] = 1.0
```

```python
a = [10,20,5,15,12.5,17.5,7.5]
mode1_analytic = [np.sqrt((5-np.sqrt(19))*9.81*1000/i) for i in a]
mode1_simulation = [17,122.9,67.6,7.53,10.8,5.6,30.1]
mode2_analytic = [np.sqrt((5+np.sqrt(19))*9.81*1000/i) for i in a]
mode2_simulation = [231.7,58,914.9,102.7,148.3,76,408.1]
```

```python
plt.figure(figsize=(9,6))
plt.plot(a,mode1_analytic,'x',label='analytic solution')
plt.plot(a,mode1_simulation,'.',label='nastran simulation')
plt.legend()
plt.xlabel('length of rod/string $a$ (mm)')
plt.ylabel('normal mode frequency $\omega$ (Hz)')
plt.title('Normal Mode 1')
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.plot(a,mode2_analytic,'x',label='analytic solution')
plt.plot(a,mode2_simulation,'.',label='nastran simulation')
plt.legend()
plt.xlabel('length of rod/string $a$ (mm)')
plt.ylabel('normal mode frequency $\omega$ (Hz)')
plt.title('Normal Mode 2')
plt.show()
```
