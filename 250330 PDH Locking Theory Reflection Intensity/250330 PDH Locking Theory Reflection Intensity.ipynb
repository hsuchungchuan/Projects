```python
def reflected_intensity(frequency_deviation, gamma):
    return 1-1/(1+(frequency_deviation/(gamma))**2)

import numpy as np

for i in range(5):
    plt.plot(np.linspace(-20,20,1000), reflected_intensity(np.linspace(-20,20,1000), i+1), label=r"$\Gamma$ = {}".format(i+1))
plt.xlabel(r"Frequency Detuning $\Delta \omega$ (a.u.)")
plt.ylabel(r"Reflected Intensity $I_{ref}$ ($I_0$)")
plt.xticks(np.arange(-20, 21, 5))
plt.legend()

plt.savefig("250330ReflectedIntensity.pdf", dpi=1200)
```
