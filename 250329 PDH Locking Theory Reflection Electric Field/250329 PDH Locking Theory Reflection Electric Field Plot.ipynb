```python
# plot two figures side by side
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.set_title("Perfect Resonance")
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Electric Field Amplitude (a.u.)")
import numpy as np
ax1.plot(np.linspace(0, 10, 100), np.sin(np.linspace(0, 10, 100)), label="Promptly reflected beam", color = "blue")
ax1.plot(np.linspace(0, 10, 100), -np.sin(np.linspace(0, 10, 100)), label="Leakage beam", color = "red")
ax1.plot(np.linspace(0, 10, 100), np.zeros(100), label=r"Total reflected field $E_{ref}$", color = "black")
ax1.legend()

ax2.set_title("Near Resonance")
ax2.set_xlabel("Time (s)")
ax2.plot(np.linspace(0, 10, 100), np.sin(np.linspace(0, 10, 100)), label="Promptly reflected beam", color = "blue")
# plot the leakage beam as the same as promptly reflected beam but phase shifted by a bit
ax2.plot(np.linspace(0, 10, 100), -np.sin(np.linspace(0, 10, 100) + np.pi/16), label="Leakage beam", color = "red")
# plot the total reflected field as the sum of the two beams
ax2.plot(np.linspace(0, 10, 100), np.sin(np.linspace(0, 10, 100)) - np.sin(np.linspace(0, 10, 100) + np.pi/16), label=r"Total reflected field $E_{ref}$", color = "black")
ax2.legend()


plt.savefig("250329ReflectedElectricFieldForDifferentLaserFrequencyRegimes.pdf", dpi=1200,pad_inches=0)
```
