```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sinusodial 

plt.plot(np.linspace(0, 50, 1000), np.sin(np.linspace(0, 50, 1000)),'k',lw=7.5)
plt.savefig("SinsudoialModulation.pdf",dpi=1200)
```

```python
# Positive triangular sawtooth plot with symmetry of 0% and 100%

plt.plot(np.linspace(0, 50, 1000), np.abs(np.mod(np.linspace(0, 50, 1000), 10)+10),'k',lw=7.5)
plt.savefig("PositiveSawtoothModulation.pdf",dpi=1200)
```
