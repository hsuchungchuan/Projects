# PDH On

```python
filename = "WavelengthDriftPDHOn.xlsx"
df = pd.read_excel(filename,header=39)
```

```python
df
```

```python
t = df.iloc[:,0]/1e3 #change from ms to seconds
t
```

```python
freq = (299792458/(df.iloc[:,1]*1e-9))/1e12
freq
```

```python
plt.figure(figsize=(9,6))
plt.plot(t,freq)
ax = plt.gca()
ax.get_yaxis().get_major_formatter().set_useOffset(False)
plt.tight_layout()
```

```python
t[81075:148857]
```

```python
plt.figure(figsize=(9,6))
plt.plot(t[81075:148857],freq[81075:148857],'.')
ax = plt.gca()
ax.get_yaxis().get_major_formatter().set_useOffset(False)
plt.tight_layout()
plt.ylabel('Frequency / THz')
plt.xlabel('Time / s')
plt.title('221108 Wavelength drift with PDH on')
# plt.savefig('221108 Wavelength drift with PDH on.png',bbox_inches='tight',dpi=300)
```
