```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the xlsx file into a pandas DataFrame
df = pd.read_excel('ShelvingIntensity_vs_Detuning_Data.xlsx')
```

```python
df
```

```python
plt.figure(figsize=(10,6))
#include error bars in y using the std dev column
plt.errorbar(df.iloc[1:,3], df.iloc[1:,4], yerr=df.iloc[1:,5], xerr=df.iloc[1:,8], fmt='o', capsize=3, capthick=1)
plt.xticks(np.arange(-25, 36, 5))
plt.title('Normalised Shelving Peak Intensity vs Frequency Detuning')
plt.xlabel('Frequency Detuning (MHz)')
plt.ylabel('Normalised Shelving Peak Intensity')
plt.grid()
# plt.savefig('ShelvingPeakIntensity_vs_Detuning.png',dpi=300)
# plt.savefig('ShelvingPeakIntensity_vs_Detuning.pdf',dpi=300)
```

```python
plt.figure(figsize=(10,6))
#include error bars in y using the std dev column
plt.errorbar(df.iloc[1:,3], df.iloc[1:,6], yerr=df.iloc[1:,7], xerr=df.iloc[1:,8], fmt='o', capsize=3, capthick=1)
plt.xticks(np.arange(-25, 36, 5))
plt.title('Normalised Shelving Dip Intensity vs Frequency Detuning')
plt.xlabel('Frequency Detuning (MHz)')
plt.ylabel('Normalised Shelving Dip Intensity')
plt.grid()
# plt.savefig('ShelvingDipIntensity_vs_Detuning.png',dpi=300)
# plt.savefig('ShelvingDipIntensity_vs_Detuning.pdf',dpi=300)
```

```python
# Find the average of column 4 and column 6 i.e. the peaks and dips

df['Average'] = (df.iloc[1:,4] + df.iloc[1:,6])/2
df['AverageError'] = (df.iloc[1:,5] + df.iloc[1:,7])/2

plt.figure(figsize=(10,6))
#include error bars in y using the std dev column
plt.errorbar(df.iloc[1:,3], df.iloc[1:,9], yerr=df.iloc[1:,10], xerr=df.iloc[1:,8], fmt='o', capsize=3, capthick=1)
plt.xticks(np.arange(-25, 36, 5))
plt.title('Normalised Shelving Average Intensity vs Frequency Detuning')
plt.xlabel('Frequency Detuning (MHz)')
plt.ylabel('Normalised Shelving Average Intensity')
plt.grid()
# plt.savefig('ShelvingAverageIntensity_vs_Detuning.png',dpi=300)
# plt.savefig('ShelvingAverageIntensity_vs_Detuning.pdf',dpi=300)
```
