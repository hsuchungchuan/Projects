```python
import numpy as np
import matplotlib.pyplot as plt

average_amplitudes_1 = np.load('shelving_amplitudes.npz')['total_average_amplitudes']
standard_deviations_1 = np.load('shelving_amplitudes.npz')['total_standard_deviations']
frequency_list_1 = np.load('shelving_amplitudes.npz')['total_frequency_lists']
print(frequency_list_1)

average_amplitudes_2 = np.load('shelving_amplitude_2.npz')['total_average_amplitudes']
standard_deviations_2 = np.load('shelving_amplitude_2.npz')['total_standard_deviations']
frequency_list_2 = np.load('shelving_amplitude_2.npz')['total_frequency_lists']
print(frequency_list_2)
# plt.errorbar(frequency_list_1, average_amplitudes_1, yerr=standard_deviations_1, fmt='o')
# plt.errorbar(frequency_list_2, average_amplitudes_2, yerr=standard_deviations_2, fmt='o')
scale_difference = average_amplitudes_2[0]-average_amplitudes_1[11]
average_amplitudes_1 = average_amplitudes_1 + scale_difference
total_average_amplitudes = np.concatenate((average_amplitudes_1, average_amplitudes_2), axis=0)
total_standard_deviations = np.concatenate((standard_deviations_1, standard_deviations_2), axis=0)
total_frequency_lists = np.concatenate((frequency_list_1, frequency_list_2), axis=0)

plt.errorbar(total_frequency_lists, total_average_amplitudes, yerr=total_standard_deviations, fmt='o')
for x, y in zip(total_frequency_lists, total_average_amplitudes):
    plt.text(x, y, str(x))
plt.xlim(210, 270)


```
