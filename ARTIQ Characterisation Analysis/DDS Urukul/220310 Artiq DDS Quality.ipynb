# 220704 Replot the Frequency Response Plots as .pdf vector format

#### Change to .pdf vector format
#### Change font size

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import time as time
plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 18
plt.rcParams['legend.fontsize'] = 18
plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['ytick.direction'] = 'out'
plt.rcParams['xtick.major.size'] = 5.0
plt.rcParams['xtick.minor.size'] = 3.0
plt.rcParams['ytick.major.size'] = 5.0
plt.rcParams['ytick.minor.size'] = 3.0
plt.rcParams['legend.handlelength'] = 1.0
```

```python
df = pd.read_csv('100322 Artiq DDS Quality.csv')

x = df.iloc[41:,2].astype(float)

y = df.iloc[41:,1].astype(float)

plt.figure(figsize=(9,6))
plt.plot(x,y,'.')
plt.yticks(np.arange(-160,10,20))
plt.xticks(np.arange(9.9994,10.0008,0.0002))
plt.ylabel('Measured power / dBm')
plt.xlabel('Frequency / MHz')
plt.title('Amplitude = 0.3, attenuation = 30, frequency = 100MHz')
# plt.show()
plt.savefig('urukul_qualitycheck'+'.pdf',bbox_inches='tight',dpi=300)
```
