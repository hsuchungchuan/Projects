**3m cable**!!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import time as time
plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 18
plt.rcParams['legend.fontsize'] = 18
plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['ytick.direction'] = 'out'
plt.rcParams['xtick.major.size'] = 5.0
plt.rcParams['xtick.minor.size'] = 3.0
plt.rcParams['ytick.major.size'] = 5.0
plt.rcParams['ytick.minor.size'] = 3.0
plt.rcParams['legend.handlelength'] = 1.0
```

```python
filename = '220728 Power vs Attn and Ampl.xlsx'
sheetname = 'sheet1'
row_start = 2
row_end = 15

df = pd.read_excel(filename,sheet_name=sheetname,header=int(row_start)-2,nrows=int(row_end)-int(row_start)+1)
df = df.sort_values('attenuation as a fct of rf power at 350 MHz')

attn = df.iloc[:,0]
rf_power = df.iloc[:,3]
plt.figure(figsize=(7.5,6))
plt.plot(attn,rf_power,'o--')
# plt.yticks(np.linspace(0,2.2,11))
plt.ylabel('RF Power / W')
plt.xlabel('Attenuation / arb.')
plt.title('Measured RF power for 350MHz (3m cable)')
# plt.show()

plt.savefig('220728 urukul rf power vs attn'+'.pdf',bbox_inches='tight',dpi=300)
```
