# 220704 Replot the Frequency Response Plots as .pdf vector format

#### Change to .pdf vector format
#### Change font size

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import time as time
plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 18
plt.rcParams['legend.fontsize'] = 18
plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['ytick.direction'] = 'out'
plt.rcParams['xtick.major.size'] = 5.0
plt.rcParams['xtick.minor.size'] = 3.0
plt.rcParams['ytick.major.size'] = 5.0
plt.rcParams['ytick.minor.size'] = 3.0
plt.rcParams['legend.handlelength'] = 1.0
```

1 

```python
df = pd.read_excel('220302 power_vs_freq.xlsx')
```

```python
x = df.iloc[:,0]
y = df.iloc[:,2]
plt.figure(figsize=(7.5,6))
plt.plot(x,y,'x',markersize=10)
plt.xlabel('Set frequency / MHz')
plt.ylabel('Measured power / dBm')
plt.title('Mirny RF Power against input frequency')
plt.savefig('mirny_power_vs_freq'+'.pdf',bbox_inches='tight',dpi=300)
```

```python
x = df.iloc[:,0]
y = df.iloc[:,1]
plt.figure(figsize=(7.5,6))
plt.plot(x,y,'x',markersize=10)
plt.plot(x,x)
plt.xlabel('Set frequency / MHz')
plt.ylabel('Measured frequency / MHz')
plt.title('Mirny measured frequency against input frequency')
plt.savefig('mirny_outfreq_vs_infreq'+'.pdf',bbox_inches='tight',dpi=300)
```

```python
df = pd.read_excel('220302 power_vs_attn.xlsx')
```

```python
x = df.iloc[:,0]
y = df.iloc[:,1]
plt.figure(figsize=(9,6))
plt.plot(x,y,'x',markersize=10)
plt.xlabel('Attenuation / bytes')
plt.ylabel('Measured power / dBm')
plt.title('Mirny measured RF Power as a function of attenuation')
```

```python
# fit linear curve

from scipy.optimize import curve_fit
```

```python
def fit_func(x,m,c):
    return m*x+c

params = curve_fit(fit_func,x[:-3],y[:-3])

[m,c]=params[0]
```

```python
m,c
```

```python
x = df.iloc[:,0]
y = df.iloc[:,1]
plt.figure(figsize=(9,6))
plt.plot(x,y,'x',markersize=10)
plt.plot(x,m*x+c,label='linear fit')
plt.xlabel('Attenuation / bytes')
plt.ylabel('Measured power / dBm')
plt.legend()
plt.title('Mirny measured RF Power as a function of attenuation')
plt.savefig('mirny_power_vs_attn'+'.pdf',bbox_inches='tight',dpi=300)
```
