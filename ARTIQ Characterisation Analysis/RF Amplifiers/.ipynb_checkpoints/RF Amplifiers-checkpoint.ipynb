# 220704 Replot the Frequency Response Plots as .pdf vector format

#### Change to .pdf vector format
#### Change font size

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import time as time
plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 18
plt.rcParams['legend.fontsize'] = 18
plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['ytick.direction'] = 'out'
plt.rcParams['xtick.major.size'] = 5.0
plt.rcParams['xtick.minor.size'] = 3.0
plt.rcParams['ytick.major.size'] = 5.0
plt.rcParams['ytick.minor.size'] = 3.0
plt.rcParams['legend.handlelength'] = 1.0
```

090322 Testing RF Amplifiers

```python
df = pd.read_excel('RF Amplifiers 220309 Data.xlsx')

p_input = df.iloc[2:14,3]
p_output = df.iloc[2:14,2]

plt.figure(figsize=(9,6))
plt.plot(p_input,'x--',label='input power')
plt.plot(p_output,'x--',label='output power')
plt.plot(p_output-p_input,'x--',label='power gain')
plt.xlabel('Run number ')
plt.ylabel('Power / dBm')
plt.title('Characterisation of RF amplifiers')
plt.legend()
# plt.show()
plt.savefig('rfamp_results'+'.pdf',bbox_inches='tight',dpi=300)
```

```python
np.average(p_output-p_input)
```

```python
from scipy.signal import find_peaks
```

```python
df = pd.read_csv('060422 DDS Urukul RF Amplifier Sideband Check 1.csv',header=42)

frequency = df.iloc[:,0]
intensity = df.iloc[:,1]

plt.figure(figsize=(7.5,6))
plt.plot(frequency,intensity)
plt.plot(frequency[np.argmax(intensity)],np.max(intensity),'rx')
plt.annotate('Output power = %.1f dBm'%np.max(intensity),[frequency[np.argmax(intensity)]+0.25e8,np.max(intensity)-1])
plt.ylabel('Intensity / dBm')
plt.xlabel('Frequency / Hz')
plt.title('Amplitude = 0.5, Attenuation = 10,\n Freq = 100MHz, Input Power = -17dBm')
# plt.show()
plt.savefig('amp1check1'+'.pdf',bbox_inches='tight',dpi=300)
```

```python
df = pd.read_csv('060422 DDS Urukul RF Amplifier Sideband Check 2.csv',header=42)

frequency = df.iloc[:,0]
intensity = df.iloc[:,1]

plt.figure(figsize=(7.5,6))
plt.plot(frequency,intensity)

plt.plot(frequency[np.argmax(intensity)],np.max(intensity),'rx') #first peak plot
plt.annotate('Output power = %.1f dBm'%np.max(intensity),[frequency[np.argmax(intensity)]+0.25e8,np.max(intensity)-1])

second_peak_index = find_peaks(intensity,height=0)[0][1] #second peak index
plt.plot(frequency[second_peak_index],intensity[second_peak_index],'rx')
plt.annotate('Output power = %.1f dbm'%intensity[second_peak_index],[frequency[second_peak_index],intensity[second_peak_index]+2])
plt.ylabel('Intensity / dBm')
plt.xlabel('Frequency / Hz')
plt.title('Amplitude = 1, Attenuation = 10,\n Freq = 100MHz, Input Power = -11dBm')
# plt.show()
plt.savefig('amp1check2'+'.pdf',bbox_inches='tight',dpi=300)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 15
plt.rcParams['legend.fontsize'] = 18
plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['ytick.direction'] = 'out'
plt.rcParams['xtick.major.size'] = 5.0
plt.rcParams['xtick.minor.size'] = 3.0
plt.rcParams['ytick.major.size'] = 5.0
plt.rcParams['ytick.minor.size'] = 3.0
plt.rcParams['legend.handlelength'] = 1.0
```

```python
df = pd.read_csv('060422 DDS Urukul RF Amplifier Sideband Check 1.csv',header=42)

frequency = df.iloc[:,0]
intensity = df.iloc[:,1]

plt.figure(figsize=(10,8))
plt.plot(frequency,intensity)
plt.plot(frequency[np.argmax(intensity)],np.max(intensity),'rx')
plt.annotate('Output power = %.1f dBm'%np.max(intensity),[frequency[np.argmax(intensity)],np.max(intensity)])
plt.ylabel('Intensity / dBm')
plt.xlabel('Frequency / Hz')
plt.title('Amplitude = 0.5, Attenuation = 10, Freq = 100MHz, Input Power = -17dBm')
plt.show()
```
