```python
axial_x_position_raw_data = [27,25,23,21,19,17,15,13,11,9]
axial_y_bfield_raw_data = [0.9,0.9,0.9,0.88,0.9,0.91,0.90,0.90,0.90,0.91]
axial_y_bfield_raw_uncertainty = [0.03,0.03,0.03,0.04,0.05,0.04,0.05,0.05,0.05,0.04]

axial_y_bfield_background = -0.5
axial_y_bfield_background_unceratinty = 0.05

axial_y_bfield_processed_data = [axial_y_bfield_raw_data[i] - axial_y_bfield_background for i in range(len(axial_y_bfield_raw_data))]
axial_y_bfield_processed_uncertainty = [np.sqrt(axial_y_bfield_raw_uncertainty[i]**2 + axial_y_bfield_background_unceratinty**2) for i in range(len(axial_y_bfield_raw_data))]

plt.errorbar(axial_x_position_raw_data, axial_y_bfield_processed_data, yerr=axial_y_bfield_processed_uncertainty, fmt='o', color='black', label='Data')
plt.ylim(0,1.5)
plt.xticks(axial_x_position_raw_data)
plt.xlabel('Horizontal Position (mm)')
plt.ylabel('Axial B-field (G)')
plt.title('Axial B-field vs. Horizontal position of gaussmeter \n (23mm center, background subtracted)')

# calculate the average axial b field processed data weighted by the uncertainty
average_axial_bfield = np.average(axial_y_bfield_processed_data, weights=[1/i for i in axial_y_bfield_processed_uncertainty])
average_axial_bfield_uncertainty = np.sqrt(1/sum([1/i**2 for i in axial_y_bfield_processed_uncertainty]))

print('Average axial B-field for 1A: ', np.round(average_axial_bfield,5), ' +/- ', np.round(average_axial_bfield_uncertainty,5))
```

Moving the gaussmeter up and down along the vertical axis

```python
vertical_vertical_position_raw_data = [0,2,4,6,8,10]
vertical_vertical_bfield_raw_data = [-1.56,-1.55,-1.55,-1.55,-1.55,-1.52]
vertical_vertical_bfield_raw_uncertainty = [i*0.01 for i in [3,3,2,6,3,3]]

vertical_vertical_bfield_background = -0.75
vertical_vertical_bfield_background_unceratinty = 0.03

vertical_vertical_bfield_processed_data = [vertical_vertical_bfield_raw_data[i] - vertical_vertical_bfield_background for i in range(len(vertical_vertical_bfield_raw_data))]
vertical_vertical_bfield_processed_uncertainty = [np.sqrt(vertical_vertical_bfield_raw_uncertainty[i]**2 + vertical_vertical_bfield_background_unceratinty**2) for i in range(len(vertical_vertical_bfield_raw_data))]

plt.errorbar(vertical_vertical_position_raw_data, vertical_vertical_bfield_processed_data, yerr=vertical_vertical_bfield_processed_uncertainty, fmt='o', color='black', label='Data')
plt.ylim(-1.5,0)
plt.xticks(vertical_vertical_position_raw_data)
plt.xlabel('Gaussmeter position along vertical axis (mm)')
plt.ylabel('B-field (G)')
plt.title('Vertical coils B-field vs. gaussmeter position along vertical axis \n (10mm is geometric centre, 0mm is vertically higher)',pad=20)

# calculate the average vertical b field processed data weighted by the uncertainty
average_vertical_vertical_bfield = np.average(vertical_vertical_bfield_processed_data, weights=[1/i for i in vertical_vertical_bfield_processed_uncertainty])
average_vertical_vertical_bfield_uncertainty = np.sqrt(1/np.sum([1/i**2 for i in vertical_vertical_bfield_processed_uncertainty]))

print('Average vertical B-field for 5A: ', np.round(average_vertical_vertical_bfield,2), ' +/- ', np.round(average_vertical_vertical_bfield_uncertainty,2))
print('Average vertical B-field for 1A: ', np.round(average_vertical_vertical_bfield/5,3), ' +/- ', np.round(average_vertical_vertical_bfield_uncertainty/5,3))
```

```python
# now average both averages weighted by their relevant uncertainties
average_vertical_bfield = np.average([average_vertical_transport_bfield, average_vertical_vertical_bfield], weights=[1/average_vertical_transport_bfield_uncertainty**2, 1/average_vertical_vertical_bfield_uncertainty**2])
average_vertical_bfield_uncertainty = np.sqrt(1/np.sum([1/average_vertical_transport_bfield_uncertainty**2, 1/average_vertical_vertical_bfield_uncertainty**2]))

print('Average vertical B-field for 5A: ', np.round(average_vertical_bfield,2), ' +/- ', np.round(average_vertical_bfield_uncertainty,2))
print('Average vertical B-field for 1A: ', np.round(average_vertical_bfield/5,3), ' +/- ', np.round(average_vertical_bfield_uncertainty/5,3))
```
