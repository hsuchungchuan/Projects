```python
df = pd.read_csv("scope_8.csv", header=33) # data for scope 9 i.e. sidebands and transmisison peak and errorr signals
scope_eight_time = df.iloc[:, 0].values * 1000
scope_eight_trans = df.iloc[:, 1].values * 100 + 2.5
scope_eight_error = df.iloc[:, 2].values +0.0015 # convert to mV
scope_eight_error = scope_eight_error * 1000 + 50 # convert to mV
```

```python
plt.plot(scope_eight_time, scope_eight_trans, label="Transmission Peaks", color="darkorange")
plt.plot(scope_eight_time, scope_eight_error, label="Error Signal", color="mediumseagreen")
plt.xlabel("Time (ms)")
plt.ylabel("Oscilloscope Voltage (mV)")
plt.legend(loc='upper left', bbox_to_anchor=(0.0, 0.5))

x_start = 0
x_end = 16
y_pos = 2.6

# # Draw double-headed arrow
# plt.annotate(
#     '', xy=(x_end, y_pos), xytext=(x_start, y_pos),
#     arrowprops=dict(arrowstyle='<->', color='black', lw=1.5)
# )
# # Place the label above the arrow
# plt.text(
#     (x_start + x_end) / 2, y_pos + 0.1, r'$\omega_{PDH}=5\,\mathrm{MHz}$',
#     ha='center', va='bottom', color='black'
# )
```

```python
df2 = pd.read_csv("scope_32.csv",header=1) # data for scope 9 i.e. sidebands and transmisison peak and errorr signals
scope_thirty_two_time = df2.iloc[:, 0].values * 1e6
scope_thirty_two_trans = df2.iloc[:, 1].values * 250 + 9
scope_thirty_two_error = df2.iloc[:, 2].values - 0.065
scope_thirty_two_error = scope_thirty_two_error * 1000 / 3 + 50
```

```python
plt.plot(scope_thirty_two_time-25, scope_thirty_two_trans, label="Transmission Peaks", color="darkorange")
plt.plot(scope_thirty_two_time-10, scope_thirty_two_error, label="Error Signal", color="mediumseagreen")
plt.xlabel(r"Time ($\mu$s)")
plt.ylabel("Oscilloscope Voltage (mV)")
plt.legend(loc='upper left', bbox_to_anchor=(0.0, 0.5))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True)

ax1.plot(scope_eight_time, scope_eight_trans, label="Transmission Peaks", color="darkorange")
ax1.plot(scope_eight_time, scope_eight_error, label="Error Signal", color="mediumseagreen")
ax1.set_xlabel("Time (ms)")
ax1.set_ylabel("Oscilloscope Voltage (mV)")
ax1.set_xticks([-20, -10, 0, 10, 20]) # set x ticks for the first plot
# ax1.legend(loc='upper left', bbox_to_anchor=(0.0, 0.5))

x_start = 0
x_end = 16
y_pos = 54

# Draw double-headed arrow
ax1.annotate(
    '', xy=(x_end, y_pos), xytext=(x_start, y_pos),
    arrowprops=dict(arrowstyle='<->', color='black', lw=1.5)
)
# Place the label above the arrow
ax1.text(
    (x_start + x_end) / 2, y_pos + 2.5, r'$\omega_{PDH}$',
    ha='center', va='bottom', color='black'
)

ax2.plot(scope_thirty_two_time-25, scope_thirty_two_trans, label="Transmission Peaks", color="darkorange")
ax2.plot(scope_thirty_two_time-10, scope_thirty_two_error, label="Error Signal", color="mediumseagreen")
ax2.set_xlabel(r"Time ($\mu$s)")
ax2.set_ylabel("Oscilloscope Voltage (mV)")
# ax2.legend(loc='upper left', bbox_to_anchor=(0.0, 0.5))

ax1.set_ylim(-5, 70)
ax2.set_ylim(-5, 70)

# plt.savefig("250604 PDHAllanDeviationCalibrationPlots.pdf", dpi=300)
```
