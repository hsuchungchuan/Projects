# Data Analysis for the Shelving Measurements
This notebook does some simple analysis for shelving data. The data were taken on 28th Thu 2024. We wrote a shelving sequence in the artiq system. We scanned the frequency of the 689nm laser and measured the shelving pattern. All the data were in the folder `ShelvingSequenceAutoRelock`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from shelving_analysis import Shelving
```

## Analyse the data from MirnyStart223MHz434829241THzEnd228MHzCSVs
Scan the Mirny frequency from 223 MHz to 228 MHz, which corresponds the red light frequeny from 434.829241 THz to 434.829246 THz. The data are saved in CSV files.

```python
# Read the data
file_1 = 'ShelvingSequenceAutoRelock\MirnyStart223MHz434829241THzEnd228MHzCSVs\
\MirnyStart223MHz434829241THzEnd228MHzCSVs_1.csv'
file_2 = 'ShelvingSequenceAutoRelock\MirnyStart223MHz434829241THzEnd228MHzCSVs\
\MirnyStart223MHz434829241THzEnd228MHzCSVs_2.csv'
file_3 = 'ShelvingSequenceAutoRelock\MirnyStart223MHz434829241THzEnd228MHzCSVs\
\MirnyStart223MHz434829241THzEnd228MHzCSVs_3.csv'
file_4 = 'ShelvingSequenceAutoRelock\MirnyStart223MHz434829241THzEnd228MHzCSVs\
\MirnyStart223MHz434829241THzEnd228MHzCSVs_4.csv'

shelving_1 = Shelving(file_1, file_2, file_3, file_4)
shelving_1.plot_shelving_graph(figure_size=(20,5))
```

```python
# List of frequencies in MHz, corresponding to different groups of peaks
frequency_lists_1 = [223, 224, 225, 226, 227]
# List of time intervals in seconds, corresponding to different groups of peaks 
# with different frequencies. The intervals are as follows:
# 0-19, 19-30, 30-42, 42-50, 50-70
time_intervals_1 = [0, 19, 30, 42, 50, 70]

average_amplitudes_1, standard_deviations_1 = \
    shelving_1.get_shelving_amplitude(time_intervals_1, frequency_lists_1, 
                                      normalised=True, method = 'subtraction')

```

```python
file_1 = 'ShelvingSequenceAutoRelock/MirnyStart233MHz434829251THzEnd237MHzCSVs/MirnyStart233MHz434829251THzEnd237MHzCSVs_1.csv'
file_2 = 'ShelvingSequenceAutoRelock/MirnyStart233MHz434829251THzEnd237MHzCSVs/MirnyStart233MHz434829251THzEnd237MHzCSVs_2.csv'
file_3 = 'ShelvingSequenceAutoRelock/MirnyStart233MHz434829251THzEnd237MHzCSVs/MirnyStart233MHz434829251THzEnd237MHzCSVs_3.csv'

shelving_5 = Shelving(file_1, file_2, file_3)

# shelving_5.plot_shelving_graph(figure_size=(20,5))

# List of frequencies in MHz, corresponding to different groups of peaks
frequency_lists_5 = [233, 234, 235, 236]
# List of time intervals in seconds, corresponding to different groups of peaks
# with different frequencies. The intervals are as follows:
# 0-19, 19-30, 30-42, 42-50, 50-70
time_intervals_5 = range(7, 53,11)

average_amplitudes_5, standard_deviations_5 = \
    shelving_5.get_shelving_amplitude(time_intervals_5, 
                                      frequency_lists_5, 
                                      upper_border = 250, 
                                      lower_border=140, 
                                      method='subtraction',
                                      error_bar=True,
                                      normalised=True,
                                      plot_average=False,
                                      plot_peak=False,
                                      plot = False)
```

# Concatenate all the results

```python
total_average_amplitudes = np.concatenate((average_amplitudes_1, average_amplitudes_2, 
                                           average_amplitudes_3, average_amplitudes_4, 
                                           average_amplitudes_5))
total_standard_deviations = np.concatenate((standard_deviations_1, standard_deviations_2,
                                            standard_deviations_3, standard_deviations_4,
                                            standard_deviations_5))
total_frequency_lists = np.concatenate((frequency_lists_1, frequency_lists_2,
                                        frequency_lists_3, frequency_lists_4,
                                        frequency_lists_5)) + 18

plt.figure(

)
plt.errorbar(total_frequency_lists, total_average_amplitudes, yerr=total_standard_deviations, fmt='o')
# for x, y in zip(total_frequency_lists, total_average_amplitudes):
   # plt.text(x, y, str(x))
plt.ylim(0.1, 0.7)
plt.xlim(240, 260)
plt.xlabel('Frequency (MHz) + 434.829 THz')
plt.ylabel('Normalised Shelving Amplitude')
plt.title('Shelving Amplitude vs Frequency')
```
