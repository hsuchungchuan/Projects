```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import time as time
# plt.rcParams['text.usetex'] = True
# plt.rcParams['font.size'] = 15
# plt.rcParams['legend.fontsize'] = 18
# plt.rcParams['xtick.direction'] = 'out'
# plt.rcParams['ytick.direction'] = 'out'
# plt.rcParams['xtick.major.size'] = 5.0
# plt.rcParams['xtick.minor.size'] = 3.0
# plt.rcParams['ytick.major.size'] = 5.0
# plt.rcParams['ytick.minor.size'] = 3.0
# plt.rcParams['legend.handlelength'] = 1.0
```

```python
t = np.linspace(0,2*np.pi,100)
```

```python
### Simplfied rabi equations

c2_sq = [(np.sin(i/2))**2 for i in t]
c1_sq = [(np.cos(i/2))**2 for i in t]
```

```python
plt.figure(figsize=(10,8))
plt.plot(t,c1_sq,label='Ground state')
plt.plot(t,c2_sq,'--',label='Excited state')
plt.ylabel('Population ratio')
plt.xlabel(r'Time / $\Omega$t')
plt.xticks(np.arange(0, 2.5*np.pi, step=0.5*np.pi), ['0',r'$\pi/2$',r'$\pi$',r'$3\pi/2$',r'$2\pi$'])  # Set text labels.
# plt.axhline(0.5,0.025,0.975,linestyle='--',color='grey')
plt.title(r'Rabi oscillations for $\Delta=0$')
plt.legend()
plt.show()
```

```python
### Full rabi equations, detuning \Delta is NOT zero

omega = 1 #let it be 1 cos time axis is in omega t
delta= np.pi
w = np.sqrt(omega**2 + delta**2)

c2_sq_full = [omega**2/w**2*(np.sin(i/2*w))**2 for i in t]
c1_sq_full = [delta**2/w**2+omega**2/w**2*(np.cos(i/2*w))**2 for i in t]
```

```python
plt.figure(figsize=(10,8))
plt.plot(t,c1_sq_full,label='Ground state')
plt.plot(t,c2_sq_full,'--',label='Excited state')
plt.ylabel('Population ratio')
plt.xlabel(r'Time / $\Omega$t')
plt.xticks(np.arange(0, 2.5*np.pi, step=0.5*np.pi), ['0',r'$\pi/2$',r'$\pi$',r'$3\pi/2$',r'$2\pi$'])  # Set text labels.
# plt.axhline(0.5,0.025,0.975,linestyle='--',color='grey')
plt.title(r'Rabi oscillations for $\Delta=\pi$')
plt.legend()
plt.show()
```
