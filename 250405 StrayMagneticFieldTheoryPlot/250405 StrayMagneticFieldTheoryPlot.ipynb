```python

# plot two figures side by side
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.set_ylabel("Magnetic field (G)")
# set the x axis label for both figures
ax1.set_xlabel("Distance (cm)")
ax2.set_xlabel("Distance (cm)")

import numpy as np
x = np.linspace(-1, 1, 101)
stray_field = 5
blue_gradient = 50
red_gradient = 10

ax1.plot(x[50:],blue_gradient*x[50:],color='blue',label="No stray field",alpha=0.25)
ax1.plot(x[:50],blue_gradient*x[:50],color='blue',linestyle='--',alpha=0.25)
ax1.plot(x[:50],-blue_gradient*x[:50],color='blue',linestyle='-',alpha=0.25)
ax1.set_title("Blue MOT")

ax1.plot(x[45:],blue_gradient*x[45:]+stray_field,color='blue',label="With stray field")
ax1.plot(x[:45],blue_gradient*x[:45]+stray_field,color='blue',linestyle='--')
ax1.plot(x[:45],-blue_gradient*x[:45]-stray_field,color='blue')
ax1.plot(x,np.abs(blue_gradient*x+stray_field),color='blue')

ax1.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
ax1.axhline(y=0, color='black', linestyle='-', linewidth=0.5)

# plot a big dot 
ax1.plot(0, 0, 'o', color='blue', markersize=5,alpha=0.3)
ax1.plot(-0.1, 0, 'o', color='blue', markersize=5,alpha=1)
# draw a horizontal arrow from (0,0) to (-0.1,0) that is vertically shifted down by 10
ax1.annotate("", xy=(-0.15, -4), xytext=(0.025, -4),
             arrowprops=dict(arrowstyle="->", color='blue', lw=1))
ax1.legend()

ax1.set_yticks(np.arange(-50, 51, 25))

ax2.plot(x[50:],red_gradient*x[50:],color='red',label="No stray field",alpha=0.25)
ax2.plot(x[:50],red_gradient*x[:50],color='red',linestyle='--',alpha=0.25)
ax2.plot(x[:50],-red_gradient*x[:50],color='red',linestyle='-',alpha=0.25)
ax2.set_title("Red MOT")

ax2.plot(x[45:],red_gradient*x[45:]+stray_field,color='red',label="With stray field")
ax2.plot(x[:45],red_gradient*x[:45]+stray_field,color='red',linestyle='--')
ax2.plot(x[:25],-red_gradient*x[:25]-stray_field,color='red')
ax2.plot(x,np.abs(red_gradient*x+stray_field),color='red')

ax2.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
ax2.axhline(y=0, color='black', linestyle='-', linewidth=0.5)

# plot a big dot 
ax2.plot(0, 0, 'o', color='red', markersize=5,alpha=0.3)
ax2.plot(-0.5, 0, 'o', color='red', markersize=5,alpha=1)
# draw a horizontal arrow from (0,0) to (-0.5,0) that is vertically shifted down by 10
ax2.annotate("", xy=(-0.55, -4), xytext=(0.025, -4),
             arrowprops=dict(arrowstyle="->", color='red', lw=1))

ax2.legend(loc='lower left')

# plt.suptitle("Shifts in magnetic field centre for the blue and red MOT from stray magentic fields")
# move the suptitle further up
# plt.subplots_adjust(top=0.85)

plt.savefig("StrayMagneticFieldTheoryPlot.pdf",dpi=1200)
```
