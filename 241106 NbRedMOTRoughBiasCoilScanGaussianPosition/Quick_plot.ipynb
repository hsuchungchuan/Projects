# Import and quick plot

```python
file_path = r"Z:\AION\Data\2024\11\06\rough_bias_coil_mikemariame_axial\000016667-MeasureRedMOT.h5"
# Open the HDF5 file to extract the relevant datasets
with h5py.File(file_path, 'r') as h5file:
    # Extracting the axis and data for plotting
    x_data = h5file['datasets/ndscan.rid_16667.points.axis_0'][:]
    current = np.linspace(5,-5,11)
    horizontal_pixel_sum = h5file['datasets/ndscan.rid_16667.points.channel_red_mot_normalised_bg_corrected_img_sum_horiz'][:]
    vertical_pixel_sum = h5file['datasets/ndscan.rid_16667.points.channel_red_mot_normalised_bg_corrected_img_sum_vert'][:]
    horizontal_camera_fit_pos_x = h5file['datasets/ndscan.rid_16667.points.channel_gaussian_fit_centre_x_horiz'][:]
    horizontal_camera_fit_pos_y = h5file['datasets/ndscan.rid_16667.points.channel_gaussian_fit_centre_y_horiz'][:]
    vertical_camera_fit_pos_x = h5file['datasets/ndscan.rid_16667.points.channel_gaussian_fit_centre_x_vert'][:]
    vertical_camera_fit_pos_y = h5file['datasets/ndscan.rid_16667.points.channel_gaussian_fit_centre_y_vert'][:]

import matplotlib.pyplot as plt

# Create a 3x2 grid of subplots
fig, axs = plt.subplots(3, 2, figsize=(10, 12))

# Make the left column share the same x-axis
for ax in axs[:, 0]:
    ax.sharex(axs[0, 0])

# Make the right column share the same x-axis
for ax in axs[:, 1]:
    ax.sharex(axs[0, 1])

axs[0,0].plot(current, horizontal_pixel_sum/10,'o-', markersize=18, linewidth=2, label='Horizontal',color='red')
axs[0,0].set_ylabel('Normalized pixel sum',fontsize=20)

axs[0,1].plot(current, vertical_pixel_sum/10,'o-', markersize=18, linewidth=2, label='Vertical',color='blue')
axs[0,1].set_ylabel('Normalized pixel sum',fontsize=20)

axs[1,0].plot(current, horizontal_camera_fit_pos_x,'o-', markersize=18, linewidth=2, label='Horizontal',color='red')
axs[1,0].set_ylabel('Gaussian fit centre position x',fontsize=20)

axs[1,1].plot(current, vertical_camera_fit_pos_x,'o-', markersize=18, linewidth=2, label='Vertical',color='blue')
axs[1,1].set_ylabel('Gaussian fit centre position x',fontsize=20)

axs[2,0].plot(current, horizontal_camera_fit_pos_y,'o-', markersize=18, linewidth=2, label='Horizontal',color='red')
axs[2,0].set_ylabel('Gaussian fit centre position y',fontsize=20)

axs[2,1].plot(current, vertical_camera_fit_pos_y,'o-', markersize=18, linewidth=2, label='Vertical',color='blue')
axs[2,1].set_ylabel('Gaussian fit centre position y',fontsize=20)

axs[2,0].set_xlabel('Axial bias coil current (A)',fontsize=20)
axs[2,1].set_xlabel('Axial bias coil current (A)',fontsize=20)
    
fig.suptitle('Nb red MOT for different axial bias coil currents',fontsize=20,x=0.55,y=1.00)

for i in range(3):
    for j in range(2):
        axs[i,j].grid(True)
        # axs[i,j].tick_params(direction='in')
        axs[i,j].legend(fontsize=20,loc='lower left')

# Adjust layout
plt.tight_layout()
# plt.show()
plt.savefig('Nb red MOT for different axial bias coil currents.pdf', dpi=1200)
```
