```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the xlsx file into a pandas DataFrame
df = pd.read_excel('ShelvingIntensity_vs_Detuning_Data.xlsx')
```

```python
# Find the average of column 4 and column 6 i.e. the peaks and dips

df['Average'] = (df.iloc[1:,4] + df.iloc[1:,6])/2
df['AverageError'] = (df.iloc[1:,5] + df.iloc[1:,7])/2
```

```python
## On your .ipynb, import this plotting_style by adding in and uncommenting the following lines:

import matplotlib.pyplot as plt
import sys
sys.path.append('C:/Users/Michael/Dropbox/AION/Projects')  # Replace with the relevant path that contains this script
from plotting_style import set_plotting_style
set_plotting_style()
```

```python
# Plot all three on the same graph

# plt.figure(figsize=(10,6))
#include error bars in y using the std dev column
plt.errorbar(df.iloc[1:,3], df.iloc[1:,4], yerr=df.iloc[1:,5], xerr=df.iloc[1:,8], fmt='o', markersize=5, capsize=3, capthick=1, label='Peak')
plt.errorbar(df.iloc[1:,3], df.iloc[1:,6], yerr=df.iloc[1:,7], xerr=df.iloc[1:,8], fmt='o', markersize=5, capsize=3, capthick=1, label='Dip')
plt.errorbar(df.iloc[1:,3], df.iloc[1:,9], yerr=df.iloc[1:,10], xerr=df.iloc[1:,8], fmt='o', markersize=5, capsize=3, capthick=1, label='Average')
plt.xticks(np.arange(-30, 31, 10))
plt.yticks(np.arange(0, 0.7, 0.1))
# plt.title('Normalised Shelving Intensity vs Frequency Detuning')
plt.xlabel('Frequency Detuning (MHz)')
plt.ylabel('Normalised Shelving Intensity')
# plt.grid()
plt.legend()
# plt.savefig('Poster_TVLBAI_ShelvingIntensity_vs_Detuning_All.png',dpi=300, bbox_inches='tight')
# plt.savefig('Poster_TVLBAI_ShelvingIntensity_vs_Detuning_All.pdf',dpi=300, bbox_inches='tight')
```
