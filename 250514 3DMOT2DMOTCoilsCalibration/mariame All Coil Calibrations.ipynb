```python
cx1 = [275,375,432,450,540,650,747,850]
#x2 = [290,348,456,484,558,686,762,817] 

cy1 = [-19,-9,-5,0,5,9,14,19]

cx1 = np.asarray(cx1)*0.0045*(1/0.5625)-4.095975296545374 #mm


#errors
c1err = [100,50,50,50,50,50,50,100]
c1err = np.asarray(c1err)*0.0045 *(1/0.5625)*0.5 
plt.plot(cx1,cy1,'ko')
plt.errorbar(cx1,cy1,xerr = c1err,ecolor='k',fmt='none')


#plot
plt.plot(cx1,cy1,'.')
coef = np.polyfit(cx1,cy1,1)
poly1d_fn = np.poly1d(coef) 

plt.plot(cx1, np.poly1d(np.polyfit(cx1,cy1, 1))(cx1),'r', label =  str("{:.2f}".format(coef[0]))+"A/mm")


plt.plot(cx1,cy1, 'ko', cx1, poly1d_fn(cx1), 'r') 
print("gradient is: " + str(coef[0]) + " A/mm")

#plt.plot(x2,y,'x')
plt.title('Coil 1')
plt.xlabel("Displacment (mm)")
plt.ylabel("Current (A)")
plt.legend()
plt.grid()
plt.savefig("9A on Coil 1")
```
