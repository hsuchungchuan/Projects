```python
import pandas as pd

# extract the data from the csv file from row 60 onwards

df = pd.read_csv("60.csv", skiprows=60, header=None)
freq = (df.iloc[:, 0]-350e6)/1e6
amplitude = df.iloc[:, 1]
plt.plot(freq,amplitude,'k')
plt.xlabel("Frequency (MHz)")
plt.ylabel("Amplitude (dBm)")
```

```python
import pandas as pd

# extract the data from the csv file from row 60 onwards

df2 = pd.read_csv("601.csv", skiprows=60, header=None)
freq2 = (df2.iloc[:, 0]-350e6)/1e3
amplitude2 = df2.iloc[:, 1]
plt.plot(freq2,amplitude2,'k')
plt.xlabel("Frequency (kHz)")
plt.ylabel("Amplitude (dBm)")
```

```python
# find the distance between two consecutive peaks

import numpy as np
from scipy.signal import find_peaks
peaks, _ = find_peaks(amplitude2,height=-30)
amplitude2[peaks],freq2[peaks]

freq_sep = []
for i in range(len(peaks)-1):
    freq_sep.append(freq2[peaks[i+1]]-freq2[peaks[i]])

print("The distance between two consecutive peaks is: %.2f +- %.2f"%(np.mean(freq_sep),np.std(freq_sep)))
```
