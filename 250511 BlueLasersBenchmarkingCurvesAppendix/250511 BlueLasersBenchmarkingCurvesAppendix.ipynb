```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
ECDL = pd.read_excel("/Volumes/MBQD_data/AION/Data/2025/02/24/Blue_Benchmarking.xlsx", sheet_name="ECDL",header=1)
ECDL_current = ECDL.iloc[:,0]
ECDL_power = ECDL.iloc[:,1]
plt.plot(ECDL_current, ECDL_power,'ko-')
plt.xlabel("461nm ECDL current (mA)")
plt.ylabel("461nm ECDL power (mW)")
plt.title("461nm ECDL Current-Power (I-P) curve")
plt.savefig("250511_ECDL_current_power_curve.pdf", dpi=300)
```
