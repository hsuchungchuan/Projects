```python
import numpy as np
import matplotlib.pyplot as plt

def errorsignal(frequency,r):
    return (2*r**2*np.sin(frequency)*(r**2-1)**2)/(1-2*r**2*np.cos(frequency)+r**4)**2

frequency = np.linspace(-5, 5, 1000)

for i in range(1,10):
    plt.plot(frequency,errorsignal(frequency,r=i/10),label='r = '+str(i/10))
plt.xlabel(r'Frequency $\omega$ (a.u.)')
plt.ylabel(r'Error signal $\epsilon(\omega)$ (V)')
plt.legend()
plt.title('Error signal as a function of frequency \n' + r'for slow modulation $\Omega$ << $\omega$')

plt.savefig('Error_signal_vs_freq.pdf', dpi=1200, bbox_inches='tight')
```
