# Zeeman splitting of fermionic strontium

## Formula is given by

$\Delta E_{ZeemanShift} = \mu_B g_f m_f B(x) $

where 

$\mu_B$ is the Bohr Magneton, $g_f$ is the Lande g-factor for hyperfine states, $m_f$ is the Zeeman magnetic substate, and $B(x)$ is the magnetic field

We consider two states:

$^1S_0$ as the ground state and $^3P_1$ as the excited state.

$^3P_1$ splits into three hyperfine states, $F=7/2$, $F=9/2$, $F=11/2$ as it has $J=1$ while $^1S_0$ only has $F=9/2$ as it has $J=0$.

## Consider the zeeman shifts for the trapping transition from 1S0 F=9/2 to 3P1 F'=11/2 (plotted on the left) and the stirring transition from 1S0 F=9/2 to 3P1 F'=9/2 (plotted on the right), see above.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
gF_11half = 3/11  # Land g-factor for strontium 87 F=11/2
gF_9half = 2/33
mu_B = 9.274009994e-24  # Bohr magneton (J/T)
hbar = 1.0545718e-34  # Reduced Planck constant (J*s)

# Nuclear spin for 87Sr
I = 9/2
J = 1
F = I+J

# Range of magnetic field strengths (in Tesla), including negative values
B = np.linspace(-1, 1, 200)  # Adjust the range as needed

# Convert energy to eV
eV_conversion = 6.242e12  # 1 Joule = 6.242e12 eV

# Calculate energy levels for F'=11/2 transition in eV
energies_11half = []
for mF in np.arange(-F, F + 1):
    energy_11half = gF_11half * mu_B * B * mF * eV_conversion
    energies_11half.append(energy_11half)

# Calculate energy levels for F'=9/2 transition in eV
energies_9half = []
for mF in np.arange(-F, F + 1):
    energy_9half = gF_9half * mu_B * B * mF * eV_conversion
    energies_9half.append(energy_9half)

# Set the figure size to (12,6) and two figures side by side with shared y axis
fig, axs = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# Plot the energy levels in the first subplot
for mF in np.arange(-F, F + 1):
    axs[0].plot(B, energies_11half[int(mF + F)], label=f'mF = {mF}')

axs[0].set_xlabel('Magnetic Field (T)')
axs[0].set_ylabel('Energy (eV)')
axs[0].set_title('Zeeman Shift in Fermionic Strontium (87Sr) for 3P1 F=11/2')

# Plot the energy levels in the second subplot
for mF in np.arange(-F, F + 1):
    axs[1].plot(B, energies_9half[int(mF + F)], label=f'mF = {mF}')

axs[1].set_xlabel('Magnetic Field (T)')
axs[1].set_title('Zeeman Shift in Fermionic Strontium (87Sr) for 3P1 F=9/2')

# Customize grid lines in both subplots
axs[0].grid(True, linestyle='--', linewidth=0.5, which='both', color='gray')
axs[1].grid(True, linestyle='--', linewidth=0.5, which='both', color='gray')

# Draw the bold vertical and horizontal lines
axs[0].axvline(0, color='black', linewidth=2, linestyle='--')
axs[1].axvline(0, color='black', linewidth=2, linestyle='--')
axs[0].axhline(0, color='black', linewidth=2, linestyle='--')
axs[1].axhline(0, color='black', linewidth=2, linestyle='--')

plt.xlabel('Magnetic Field (T)')
plt.ylabel('Energy (eV)')
plt.show()
```

Now we look at which zeeman states are for $\sigma^+$ transition

```python
# Set the figure size to (8,6)
plt.figure(figsize=(8, 6))

# Plot the energy levels
for mF in np.arange(-F+2, F + 1):
    plt.plot(B, energies_11half[int(mF + F)], label=f'mF = {mF}')

# Draw the bold vertical and horizontal lines
plt.axvline(0, color='black', linewidth=2, linestyle='--')
plt.axhline(0, color='black', linewidth=2, linestyle='--')
plt.xlabel('Magnetic Field (T)')
plt.ylabel('Energy (eV)')
plt.title('Zeeman Shift of relevant $\sigma^+$ transitions for 87Sr F=11/2')
plt.legend()
plt.grid(True)
plt.show()
```

Now let's repeat the same for $\sigma^-$ transition.

```python
# Set the figure size to (8,6)
plt.figure(figsize=(8, 6))

# Plot the energy levels
for mF in np.arange(-F, F + 1 -2):
    plt.plot(B, energies_11half[int(mF + F)], label=f'mF = {mF}')

# Draw the bold vertical and horizontal lines
plt.axvline(0, color='black', linewidth=2, linestyle='--')
plt.axhline(0, color='black', linewidth=2, linestyle='--')
plt.xlabel('Magnetic Field (T)')
plt.ylabel('Energy (eV)')
plt.title('Zeeman Shift of relevant $\sigma^-$ transitions for 87Sr F=11/2')
plt.legend()
plt.grid(True)
plt.show()
```
