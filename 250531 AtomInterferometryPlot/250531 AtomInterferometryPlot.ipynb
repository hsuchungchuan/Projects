```python
import numpy as np
import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, constrained_layout=True)

# Define parameters for the interference fringe
phi = np.linspace(0, 2 * np.pi, 500)  # Laser phase from 0 to 2pi
# contrast = 0.85  # Fringe contrast (C)
phase_offset = 0  # Phase offset (phi_0)

# Population asymmetry A(phi) = C * sin(phi + phi_0)
def PopulationAsymmetry(contrast):
    phi = np.linspace(0, 2 * np.pi, 500)  # Laser phase from 0 to 2pi
    # contrast = 0.85  # Fringe contrast (C)
    phase_offset = 0  # Phase offset (phi_0)
    return contrast * np.sin(phi + phase_offset)

ax2.set_xlabel(r'Laser phase (rad)')
ax1.set_ylabel(r'Population asymmetry $A$')

# Create the plot
ax1.plot(phi, PopulationAsymmetry(1), label=r'C=1.0',color='black')
ax2.plot(phi, PopulationAsymmetry(0.5), label=r'C=0.5',color='black')
ax3.plot(phi, PopulationAsymmetry(0.25), label=r'C=0.25',color='black')

# Set the x-axis limits
ax1.set_xlim(0, 2 * np.pi)
# Set the y-axis limits
ax1.set_ylim(-1.1, 1.1)

ax1.legend(loc='upper right')
ax2.legend(loc='upper right')
ax3.legend(loc='upper right')

ax1.set_xticks(np.arange(0, 2.5*np.pi, step=0.5*np.pi), ['0',r'$\pi/2$',r'$\pi$',r'$3\pi/2$',r'$2\pi$'])
ax2.set_xticks(np.arange(0, 2.5*np.pi, step=0.5*np.pi), ['0',r'$\pi/2$',r'$\pi$',r'$3\pi/2$',r'$2\pi$'])
ax3.set_xticks(np.arange(0, 2.5*np.pi, step=0.5*np.pi), ['0',r'$\pi/2$',r'$\pi$',r'$3\pi/2$',r'$2\pi$'])

plt.savefig("250531 InterferometryPlotPopulationAsymmetry.pdf", dpi=300)
```
