# 220704 Replot the Frequency Response Plots as .pdf vector format

#### Change to .pdf vector format
#### Change font size

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import time as time
plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 18
plt.rcParams['legend.fontsize'] = 18
plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['ytick.direction'] = 'out'
plt.rcParams['xtick.major.size'] = 5.0
plt.rcParams['xtick.minor.size'] = 3.0
plt.rcParams['ytick.major.size'] = 5.0
plt.rcParams['ytick.minor.size'] = 3.0
plt.rcParams['legend.handlelength'] = 1.0
```

```python
# kim data

filename = '220525 Beam Divergence Measurements.xlsx'
sheetname = '220525 kimdata_optimized'
row_start = 2
row_end = 8

df = pd.read_excel(filename,sheet_name=sheetname,header=int(row_start)-2,nrows=int(row_end)-int(row_start)+1)

df

v_r_data = df.iloc[:,4]
h_r_data = df.iloc[:,5]
v_r_fitpredict = df.iloc[:,7]
h_r_fitpredict = df.iloc[:,8]
d = df.iloc[:,1] #distance from LASER, not waist

plt.figure(figsize=(9,6))
plt.plot(d,v_r_data,'x',markersize=8,label=r'Vertical')
plt.plot(d,h_r_data,'x',markersize=8,label=r'Horizontal')
plt.plot(d,v_r_fitpredict,'--',label=r'Vertical fit')
plt.plot(d,h_r_fitpredict,'--',label=r'Horizontal fit')
plt.legend()
plt.xlabel('Distance from laser / mm')
plt.ylabel(r'Beam radius / $\mu$m')
plt.title('Beam propagation fitting, 2022-05-25')
# plt.show()
plt.savefig('220525 beam prop least sq fit kimdata'+'.pdf',bbox_inches='tight',dpi=300)
```

```python
# rezonator fitting plot using mike param 

filename = '180622 rezonator beam propagation.xlsx'
sheetname = '220618 beam prop tele AOM'
row_start = 2
row_end = 299

df = pd.read_excel(filename,sheet_name=sheetname,header=int(row_start)-2,nrows=int(row_end)-int(row_start)+1)

v_r_fitpredict = df.iloc[:,1]
h_r_fitpredict = df.iloc[:,2]
d = df.iloc[:,0] #distance from LASER, not waist

plt.figure(figsize=(9,6))
plt.plot(d,v_r_fitpredict,'r',markersize=2.5,label=r'Vertical')
plt.plot(d,h_r_fitpredict,'g',markersize=2.5,label=r'Horizontal')
plt.plot(d[266],v_r_fitpredict[266],'rx',markersize=15)
plt.plot(d[266],h_r_fitpredict[266],'gx',markersize=15)
plt.annotate(r'r$_{V}$ = %.1f$\mu$m'%v_r_fitpredict[266],[d[266],v_r_fitpredict[266]],xytext=[d[266]-150,v_r_fitpredict[266]+100])
plt.annotate(r'r$_{H}$ = %.1f$\mu$m'%h_r_fitpredict[266],[d[266],h_r_fitpredict[266]],xytext=[d[266]-150,h_r_fitpredict[266]-100])
plt.axvline(x=d[266],ymax=0.80,linestyle='--')
plt.annotate('AOM',xy=[d[266]-50,1275])
plt.legend(loc='upper left')
plt.xlabel('Propagation distance from waist / mm')
plt.ylabel(r'Beam radius / $\mu$m')
plt.title('Beam propagation fitting after telescope via Rezonator, 2022-06-18')
# plt.show()
plt.savefig('rezonator telescope'+'.pdf',bbox_inches='tight',dpi=300)
```
