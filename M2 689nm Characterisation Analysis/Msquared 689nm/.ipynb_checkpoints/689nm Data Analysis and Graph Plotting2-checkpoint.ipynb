```python
# Power Curve + internal photodiode

filename = 'msquared 689 characterization measurements.xlsx'
sheetname = '050522'
row_start = 2
row_end = 10

df = pd.read_excel(filename,sheet_name=sheetname,header=int(row_start)-2,nrows=int(row_end)-int(row_start)+1)

pumppower = df.iloc[:,0]
highpower = df.iloc[:,1]
wavelength = df.iloc[:,2]
voltagepd = df.iloc[:,3]
datetime = df.iloc[:,4] 
```

```python
df
```

```python
datetime
```

```python
time_hm = datetime.dt.strftime("%H:%M")
time_hm #time in hours and minutes
```

```python
fig, ax1 = plt.subplots(figsize=(10,8))
ax1.plot(pumppower,highpower,'bo',label='Output power')
ax2 = ax1.twinx()
ax2.plot(pumppower,voltagepd,'ro',alpha=0.5,label='PD voltage')

ax1.set_ylabel('Output power (high power PM) / W')
ax1.set_xlabel('Pump power / W')
ax2.set_ylabel('Solstis internal diode output PD / V')
datetime = time.strftime("%Y-%m-%d")
plt.title('Power curve and internal photodiode voltage ' + datetime)
fig.legend(loc='upper left',bbox_to_anchor=(0.15,0.85))
plt.show()
```

```python
fig, ax = plt.subplots(3, sharex=True,figsize=(10,10))
date = time.strftime("%Y-%m-%d")
fig.suptitle('Power, Photodiode Voltage and Wavelength fluctuations in an hour, ' + date)

ax[0].plot(time_hm, highpower,'bo',label='Output power')
ax[0].set_ylabel('Output Power / W')

ax[1].plot(time_hm, voltagepd, 'ro',alpha=0.75,label='PD voltage')
ax[1].set_ylabel('PD Voltage / V')

ax[2].plot(time_hm, wavelength, 'go',alpha=0.75,label='Wavelength')
ax[2].set_ylabel('Wavelength / nm')

plt.xticks(rotation=90)
plt.show()
```
