```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load the PNG image
image = Image.open('REDMOT_CROPPED_GREYSCALE.png').convert("L")

# Convert the image to a NumPy array
image_array = np.array(image)

# Resize the images to 1300x1300
image_resized = image.resize((1200, 1200), Image.LANCZOS)

# Convert the resized image to a NumPy array
image_resized_array = np.array(image_resized)

# Display the resized image and its shape
plt.imshow(image_resized_array)
print("Resized image shape:", image_resized_array.shape)
```
