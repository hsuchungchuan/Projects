```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
start_line = 59
df = pd.read_csv("350MHz VCO.csv" , skiprows=range(1, start_line))
t = df.iloc[:,0]/1e6
p = df.iloc[:,1]
plt.plot(t,p)
plt.title('Spectrum Analyzer Graph')
plt.ylabel('Power / dBm')
plt.xlabel('Frequency / MHz')
plt.savefig('SpectrumAnalyzerGraph.png',dpi=600)
```
