```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import time as time
```

```python
plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 18
plt.rcParams['legend.fontsize'] = 12
plt.rcParams['xtick.direction'] = 'out'
plt.rcParams['ytick.direction'] = 'out'
plt.rcParams['xtick.major.size'] = 5.0
plt.rcParams['xtick.minor.size'] = 3.0
plt.rcParams['ytick.major.size'] = 5.0
plt.rcParams['ytick.minor.size'] = 3.0
plt.rcParams['legend.handlelength'] = 1.0
```

# scope_0: 10ms

```python
df = pd.read_csv('scope_0_PDHon_10ms.csv',header=1) # scope_14

#zoom in
df;

plt.figure(figsize=(9,6))
plt.plot(df['second'],df['Volt'],label='Fast Out')
plt.plot(df['second'],df['Volt.1'],label='Transmission (Fast PD)')
plt.plot(df['second'],df['Volt.2'],label='Error Signal (Monitoring Output)')
plt.plot(df['second'],df['Volt.3'],label='Slow Out')

plt.legend()
plt.title('221109 Oscilloscope Signal, engaged PDH lock, 10ms resolution')
plt.axhline(0,color='black',ls='--')
plt.xlabel('Time / s')
plt.ylabel('Voltage / V')
plt.savefig('221109 Oscilloscope Signal, engaged PDH lock, 10ms resolution.pdf',bbox_inches='tight',dpi=300)
```
