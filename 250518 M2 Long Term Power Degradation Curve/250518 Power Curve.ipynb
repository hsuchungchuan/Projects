```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Given data
outputPD = np.array([0.302, 0.382, 0.433, 0.48, 0.533, 0.605, 0.67])
power = np.array([1.39, 1.77, 2.02, 2.27, 2.52, 2.85, 3.14])

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(outputPD, power)

# Equation of the line
line = slope * outputPD + intercept

# Plotting the data and the fitted line
plt.figure(figsize=(6, 4))
plt.scatter(outputPD, power, label="Data points", color="blue")
plt.plot(outputPD, line, label=f"Fit: Power = {slope:.3f}  OutputPD + {intercept:.3f}", color="red")
plt.xlabel("Output PD (V)")
plt.ylabel("Power (mW)")
plt.title("Calibration: Power vs. Output PD")
plt.legend()
plt.grid(True)
plt.show()

# Results
slope, intercept, r_value**2
```

```python
fig, ax = plt.subplots()

voltage_output_err = 0.005 * np.ones(len(voltage_output))
power_output_err = 0.02 * np.ones(len(power_output))

ax.errorbar(pump_power, power_output, yerr=power_output_err,
            color="crimson", 
            marker="o",
            ecolor="black",
            capsize=3,
            capthick=1,
            linestyle="-",
            label="Power Output")
ax.set_ylabel("Output Power (W)")
# Custom legend position: set bbox_to_anchor and loc as desired
ax.legend(bbox_to_anchor=(0.2425, 0.94))
ax.set_yticks([0,1,2,3,4,5])
ax.set_xlabel("Pump Power (W)")

ax2 = ax.twinx()
ax2.errorbar(pump_power, voltage_output, yerr=voltage_output_err, 
             color="dodgerblue", 
             marker="o",
             ecolor="black",
             capsize=3,
             capthick=1,
             linestyle="-",
             label="Photdiode Voltage Output", )
ax2.set_xlabel("Pump Power (W)")
ax2.set_ylabel("SolsTiS Internal Photodiode Voltage (V)")
ax2.set_yticks(np.arange(0, 1.1, 0.2))
ax2.grid(False)
ax2.legend()

plt.savefig("250518 PowerOutputVoltageVsPumpPower.pdf", dpi=300)
```
