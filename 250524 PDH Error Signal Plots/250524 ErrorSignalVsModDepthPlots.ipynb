### Fix amplitude @ 5dBm, vary frequency from 1MHz, 2MHz, 3MHz

```python
freq = 1

for i in [6,3,4]:
    print('scope_'+str(i))
    df = pd.read_csv('scope_'+str(i)+'.csv',header=1)
    plt.figure(figsize=(9,6))
    plt.plot(df['second'],df['Volt.1'],label='Transmission signal f='+str(freq)+'MHz')
    plt.plot(df['second'],df['Volt.2'],label='Error signal f='+str(freq)+'MHz',alpha=0.5)
    plt.legend()
    plt.xlabel('Time / s')
    plt.ylabel('Voltage / V')
    # plt.savefig('221014 Error Signal Ampl 5dBm Freq '+str(freq)+'MHz.png',bbox_inches='tight',dpi=300)
    freq += 1
```
