```python
phase_diff = np.array([70,60,50,40,30,20,10,0,80,90,100,110,120,130,140,150,160,170,180])
error_signal = np.array([2.32,2.25,2.15,2,1.81,1.39,1.21,0.82,2.17,2.21,2.1,1.95,1.71,1.43,1.13,0.78,0.46,0.4,0.7])
error_signal_err = np.ones(len(error_signal)) * 0.05

sorted_indices = np.argsort(phase_diff)
phase_diff = phase_diff[sorted_indices]
error_signal = error_signal[sorted_indices]
error_signal_err = error_signal_err[sorted_indices]

plt.errorbar(phase_diff/360, error_signal, yerr=error_signal_err, fmt='ko-', color='crimson',
             markeredgewidth=1.5, markeredgecolor='black', ecolor='black',capsize=3, label='Error Signal')

plt.xlabel(r'Phase Difference ($\pi$)')
plt.ylabel(r'Error Signal (V)')
plt.savefig('250524 ErrorSignalVsPhaseDifferencePlot.pdf', dpi=300)
```
